```python
import pandas as pd

# Cargar el archivo CSV
file_path = '/content/historical_automobile_sales.csv'  # Asegúrate de poner la ruta correcta aquí
df = pd.read_csv(file_path)

# Mostrar las primeras filas del dataframe para inspeccionar los datos
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Agrupar por año y calcular la suma de las ventas de automóviles por año
sales_by_year = df.groupby('Year')['Automobile_Sales'].sum().reset_index()

# Crear el gráfico de líneas
plt.figure(figsize=(10, 6))
plt.plot(sales_by_year['Year'], sales_by_year['Automobile_Sales'], marker='o')
plt.title('Fluctuación de las ventas de automóviles por año', fontsize=16)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Ventas de Automóviles', fontsize=12)
plt.grid(True)
plt.show()
```

```python


# Group the data by Year and calculate the average Automobile Sales
df_grouped_by_year = df.groupby('Year')['Automobile_Sales'].mean()

# Create the line chart
plt.figure(figsize=(10,6))
df_grouped_by_year.plot(kind='line')

# Add title and labels
plt.title('Automobile Sales during Recession')
plt.xlabel('Year')
plt.ylabel('Average Automobile Sales')

# Set x-axis ticks for all years from 1980 to 2023
plt.xticks(range(1980, 2024), rotation=45)

# Add annotation for recession periods (1980 and 2008)
plt.text(1980, df_grouped_by_year.loc[1980] + 500, 'Recession (1980)', fontsize=10, color='red')
plt.text(2008, df_grouped_by_year.loc[2008] + 500, 'Recession (2008)', fontsize=10, color='red')

# Add grid
plt.grid(True)

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filtrar los datos para recesión y no recesión
df_recession = df[df['Recession'] == 1]
df_no_recession = df[df['Recession'] == 0]

# Agrupar por tipo de vehículo y calcular la media de las ventas de automóviles
sales_by_type_recession = df_recession.groupby('Vehicle_Type')['Automobile_Sales'].mean().reset_index()
sales_by_type_non_recession = df_no_recession.groupby('Vehicle_Type')['Automobile_Sales'].mean().reset_index()

# Crear un gráfico de barras para comparar las tendencias de ventas
plt.figure(figsize=(10, 6))
sns.barplot(x='Vehicle_Type', y='Automobile_Sales', data=sales_by_type_recession, color='salmon', label='Recesión', alpha=0.6)
sns.barplot(x='Vehicle_Type', y='Automobile_Sales', data=sales_by_type_non_recession, color='skyblue', label='No Recesión', alpha=0.6)

# Personalizar el gráfico
plt.title('Ventas de automóviles por tipo de vehículo en recesión y no recesión', fontsize=16)
plt.xlabel('Tipo de Vehículo', fontsize=12)
plt.ylabel('Ventas de Automóviles', fontsize=12)
plt.legend(title='Periodo')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filtrar los datos para recesión y no recesión
df_recession = df[df['Recession'] == 1]
df_no_recession = df[df['Recession'] == 0]

# Agrupar por año y calcular la media del PIB para recesión y no recesión
gdp_recession = df_recession.groupby('Year')['GDP'].mean().reset_index()
gdp_no_recession = df_no_recession.groupby('Year')['GDP'].mean().reset_index()

# Crear los subgráficos para comparar las variaciones del PIB
fig, axes = plt.subplots(1, 2, figsize=(15, 6), sharey=True)

# Trazado de PIB durante recesión
axes[0].plot(gdp_recession['Year'], gdp_recession['GDP'], color='red', marker='o', label='Recesión')
axes[0].set_title('Variación del PIB en periodo de Recesión', fontsize=14)
axes[0].set_xlabel('Año', fontsize=12)
axes[0].set_ylabel('PIB Promedio', fontsize=12)
axes[0].grid(True)
axes[0].legend()

# Trazado de PIB durante no recesión
axes[1].plot(gdp_no_recession['Year'], gdp_no_recession['GDP'], color='blue', marker='o', label='No Recesión')
axes[1].set_title('Variación del PIB en periodo de No Recesión', fontsize=14)
axes[1].set_xlabel('Año', fontsize=12)
axes[1].grid(True)
axes[1].legend()

# Título general para ambos subgráficos
plt.suptitle('Comparación del PIB entre Recesión y No Recesión', fontsize=16)

# Mostrar los gráficos
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filtrar los datos para la visualización
df_bubble = df[['Year', 'Automobile_Sales', 'Seasonality_Weight']]

# Crear un gráfico de burbujas
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_bubble, x='Year', y='Automobile_Sales', size='Seasonality_Weight',
                sizes=(20, 200), hue='Seasonality_Weight', palette='coolwarm', legend=None)

# Personalizar el gráfico
plt.title('Impacto de la Estacionalidad en las Ventas de Automóviles', fontsize=16)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Ventas de Automóviles', fontsize=12)
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
temp = df.copy()
months=temp["Month"].unique().tolist()
# display(months)
temp["Month"] = temp["Month"].replace(dict(zip(months, range(0,len(months)))))
temp = temp.groupby("Month")[["Automobile_Sales", "Seasonality_Weight"]].mean().reset_index()
# display(temp)
plt.figure(figsize=(10, 6))
plt.xticks(ticks=range(0,len(months)),labels=months)
sns.scatterplot(data=temp,
                x='Month',
                y='Automobile_Sales',
                hue="Seasonality_Weight",
                size="Seasonality_Weight",  # pass in weights
                #sizes=(20, 1500)
               )
plt.xlabel("Month")
plt.ylabel("Automobile Sales")
plt.title("Automobile Sales per Month according to Seasonality Weight")
plt.legend(loc="right")
```

```python
import matplotlib.pyplot as plt

# Filtrar los datos para los periodos de recesión
df_recession = df[df['Recession'] == 1]

# Crear un gráfico de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(df_recession['Price'], df_recession['Automobile_Sales'], color='b', alpha=0.6)

# Personalizar el gráfico
plt.title('Correlación entre el Precio y las Ventas de Automóviles durante la Recesión', fontsize=16)
plt.xlabel('Precio Medio de los Vehículos (USD)', fontsize=12)
plt.ylabel('Ventas de Automóviles', fontsize=12)
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
import plotly.express as px

# Filtrar los datos para los periodos de recesión y no recesión
df_recession = df[df['Recession'] == 1]
df_no_recession = df[df['Recession'] == 0]

# Calcular el total de la publicidad durante los periodos de recesión y no recesión
advertising_recession = df_recession['Advertising_Expenditure'].sum()
advertising_no_recession = df_no_recession['Advertising_Expenditure'].sum()

# Crear el gráfico circular
labels = ['Recesión', 'No Recesión']
values = [advertising_recession, advertising_no_recession]

fig = px.pie(values=values, names=labels, title='Distribución del Gasto en Publicidad durante la Recesión y No Recesión')

# Mostrar el gráfico
fig.show()
```

```python
# Filtrar los datos para el periodo de recesión
df_recession = df[df['Recession'] == 1]

# Agrupar los datos por tipo de vehículo y calcular el total del gasto en publicidad
advertising_by_vehicle_type = df_recession.groupby('Vehicle_Type')['Advertising_Expenditure'].sum().reset_index()

# Crear el gráfico circular
fig = px.pie(advertising_by_vehicle_type,
             values='Advertising_Expenditure',
             names='Vehicle_Type',
             title='Gasto Total en Publicidad por Tipo de Vehículo durante la Recesión')

# Mostrar el gráfico
fig.show()
```

```python
# Filtrar los datos para el periodo de recesión
df_recession = df[df['Recession'] == 1]

# Agrupar los datos por tipo de vehículo y tasa de desempleo
sales_by_unemployment = df_recession.groupby(['Vehicle_Type', 'unemployment_rate'])['Automobile_Sales'].sum().reset_index()

# Crear el gráfico de líneas
plt.figure(figsize=(10, 6))
for vehicle in sales_by_unemployment['Vehicle_Type'].unique():
    vehicle_data = sales_by_unemployment[sales_by_unemployment['Vehicle_Type'] == vehicle]
    plt.plot(vehicle_data['unemployment_rate'], vehicle_data['Automobile_Sales'], marker='o', label=vehicle)

plt.title('Efecto de la Tasa de Desempleo sobre las Ventas de Automóviles por Tipo de Vehículo durante la Recesión', fontsize=16)
plt.xlabel('Tasa de Desempleo', fontsize=12)
plt.ylabel('Ventas de Automóviles', fontsize=12)
plt.legend(title='Tipo de Vehículo')
plt.grid(True)
plt.show()
```
